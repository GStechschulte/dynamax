### Define the STS model

```python
# The model includes the local linear trend component and the seasonal component
trend = sts.LocalLinearTrend(observed_timeseries=co2_by_month_training_data)
seasonal = sts.Seasonal(num_seasons=12, observed_timeseries=co2_by_month_training_data)
model = sts.StructuralTimeSeries([trend, seasonal], observed_timeseries=co2_by_month_training_data)
```

### Model fitting

```python
# Fit the model using HMC
observed_time_series = co2_by_month_training_data
batch_observed_time_series = jnp.array([co2_by_month_training_data])
key = jr.PRNGKey(0)
sample_size = 200

parameter_samples = model.fit_hmc(key, sample_size, batch_observed_time_series, 
                                  batch_inputs=None, warmup_steps=500, num_integration_steps=30)
```

### Forecast

```python
predicts = model.forecast(key, observed_time_series, parameter_samples, num_forecast_steps)
pred_means = jnp.squeeze(predicts['means']).mean(axis=0)
pred_scales = jnp.sqrt(jnp.squeeze(predicts['covariances'])).mean(axis=0)

# Plot the prediction
fig = plt.figure(figsize=(12, 6))
ax = fig.add_subplot(1, 1, 1)
ax.plot(co2_dates, co2_by_month, lw=2, label='real observation')
ax.plot(co2_dates[-num_forecast_steps:], pred_means, lw='2', color='orange', label='prediction')
ax.fill_between(co2_dates[-num_forecast_steps:], 
                pred_means-2*pred_scales, pred_means+2*pred_scales, color='orange', alpha=0.3)
ax.set_ylabel("Atmospheric CO2 concentration (ppm)")
ax.set_xlabel("Year")
fig.suptitle("Monthly average CO2 concentration Mauna Loa, Hawaii", fontsize=15)
```
